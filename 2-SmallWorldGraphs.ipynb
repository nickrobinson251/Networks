Begin with a set of nodes that are identified with the set of lattice points in an $n \times n$ square, $$\{(i, j): i \in \{1, 2, \ldots, n\}, j \in \{1, 2, \ldots, n\}\}$$ 

Define the _lattice distance_ between two nodes $(i, j)$ and $(k, l)$ to be the number of "lattice steps" separating them: $$d((i, j), (k, l)) = |k - i| + |l - j|$$

      
For a universal constant $p \geq 1$, the node $u$ has a directed edge to every other node within lattice distance $p$ --- these are its *local
contacts*. 

For universal constants $q \geq 0$ and $r \geq 0$, construct directed edges from $u$ to $q$ other nodes (the *long-range contacts*) using independent random trials; the $i^{th}$ directed edge from $u$ has endpoint $v$ with probability proportional to $[d(u,v)]^{-r}$.
